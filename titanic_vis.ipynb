# Titanic Visualization

## 1.  Barcharts

```r
p1 = function(y)
{
  library(ggplot2)
  rt = ggplot(data = y)+geom_bar(aes(x = y[[1]], fill = y[[2]]), 
                                 position = 'dodge')+ xlab(names(y)[1]) + labs(fill = names(y)[2])
  return(rt)
}


quick_visual = function(df)
{
  # get the list of all "small" categorical variables
  d1 = df[,apply(df, 2, function(x){length(table(x))<5})]
  id = combn(ncol(d1), 2)

  pl = apply(id, 2, function(x){p1(d1[,x])})

  return(pl)
}
```

```r
library(ggplot2)
titanic = read.csv("C:/Users/snguyen4/Dropbox/1. Teaching/1. Fall 2018/Math 421/train.csv")

# Remove some columns
titanic$PassengerId =  NULL
titanic$Ticket =  NULL
titanic$Name = NULL
titanic$Cabin = NULL

# Correct variables' types
titanic$Survived = factor(titanic$Survived)
titanic$Pclass = factor(titanic$Pclass)
```

```r
quick_visual(titanic)
```
